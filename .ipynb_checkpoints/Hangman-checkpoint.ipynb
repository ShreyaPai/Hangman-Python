```python
import random
playAgain="y"

def guessWord():
    wordsList = ['coffee', 'lemonade', 'diwali', 'christmas','january', 'august','jackal', 'oyseter',
                 'ostrich', 'tiger','hammer', 'elephant', 'pigeon', 'flamingo', 'pencil','giraffe', 
                 'window', 'bison', 'sugar', 'pepper','antenna','hoist', 'freedom', 'sports', 'summer',
                 'winter', 'laptop', 'jacket', 'jasmine', 'tulips', 'cucumber', 'yacht','sneakers',
                 'orchids','arrogant', 'joker', 'mango', 'grapes', 'tomato', 'potato', 'carrot','mirror'
                 'mosquito', 'frame', 'salsa', 'pepper', 'rhinos', 'aeroplane', 'stars', 'india',
                 'china','malasia', 'trust', 'reliable', 'health', 'kettle', 'kenya','australia'] 
    
    botWord = random.choice(wordsList)
    botWord = botWord.strip()
    return botWord

def printHangman(guesses, word):
    #guesses = 0
    if guesses <= 6:
        if(guesses==0):
            print("H\n")
        elif(guesses==1):
            print("HA\n")
        elif(guesses==2):
            print("HAN\n")
        elif(guesses==3):
            print("HANG\n")
        elif(guesses==4):
            print("HANGM")
        elif(guesses==5):
            print("HANGMA")
        elif(guesses==6):
            print("HANGMAN! You lose")
            print(word+" was the word\n")
        else:
            print("YAY you win!")
    return
        
def beginGame():
    guesses = 0
    word=guessWord()
    word_list=list(word)
    dash="*"*len(word)
    dashList=list(dash)
    guessList = []
    print("Begin\n")
    printHangman(guesses, word)
    print(' '.join(dashList))
    
    while guesses <= 5:
        userInput=input("\nGuess Letter\n")
        userInput=str(userInput)
        
        if len(userInput) > 1:
            print("You can enter only one letter")
        elif len(userInput) < 1:
            print("Empty")
        elif userInput in guessList:
            print("Letter already exits!")
        else:
            guessList.append(userInput)
            if userInput in word_list:
                #print("Great going! Guess again")
                for i in range(0, len(word_list)): 
                    if userInput == word[i]:
                        print("Right guess!")
                        dashList[i] = userInput
                        print(dashList)
                        
                if not '*' in dashList:
                            print("Congratulations, You won!\n")
                            print(word+" was the word")
                            break
                
            else:
                print("Wrong guess! Try again")
                guesses = guesses + 1
                printHangman(guesses,word)
                       
beginGame()
 
while playAgain == 'y':
    print("Do you want to play again? y or n")
    playAgain=input('>')
    playAgain=str(playAgain)
    if playAgain == 'y':
        beginGame()
    if playAgain == 'n':
        exit
```
